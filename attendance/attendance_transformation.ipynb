```python
#import dependencies
import pandas as pd

#create file path variable
nfl = "../../Data/ETL_project_data/nfl_attendance.csv"
mlb = "../../Data/ETL_project_data/mlb_attendance.csv"
nba = "../../Data/ETL_project_data/nba_attendance.csv"
nhl = "../../Data/ETL_project_data/nhl_attendance.csv"
cities = "../../Data/ETL_project_data/Cities_SportsTeams.csv"

#read .csv files
nfl_df = pd.read_csv(nfl)
mlb_df = pd.read_csv(mlb)
nba_df = pd.read_csv(nba)
nhl_df = pd.read_csv(nhl)
cities_df = pd.read_csv(cities)

nfl_df.head()
```

```python
#create column to identify sport
nfl_df["LEAGUE"] = "NFL"
mlb_df["LEAGUE"] = "MLB"
nba_df["LEAGUE"] = "NBA"
nhl_df["LEAGUE"] = "NHL"


#drop extra totals rows in NFL csv
nfl_df = nfl_df.drop(columns=['TOTAL.1','TOTAL.2'])

nfl_df.head()
```

```python
#concatenate attendance dataframes into one dataframe
attendance_df = pd.concat([nfl_df,mlb_df,nba_df,nhl_df])
attendance_df.head()
```

```python
cities_df.head()
```

```python
new_city = cities_df[["City", "NFL", "MLB", "NBA", "NHL"]]
```

```python

teams = {}

for i in ["NFL", "MLB", "NBA", "NHL"]:
    sport = new_city[i]
    for k,j in enumerate(sport):
        teams[j] = new_city['City'].iloc[k]
```

```python
def getcity(x):
    return teams[x] if x in teams.keys() else ''
```

```python
attendance_df['CITY'] = attendance_df['TEAM'].apply(getcity)
```

```python
#drop RANK and TOTAL columns
attendance_df = attendance_df.drop(columns=['YEAR_RANK','TOTAL'])
```

```python

#rename dataframe columns
attendance_df = attendance_df.rename( columns = {
    "AVG":"HOME_AVG", "PCT":"HOME_PCT", "AVG.1":"ROAD_AVG", "PCT.1":"ROAD_PCT", "OVERALL_GMS":"OVR_GMS", 
    "AVG.2":"OVR_AVG", "PCT.2":"OVR_PCT"
})

#re-order columns
attendance_df = attendance_df[['CITY','TEAM','LEAGUE','YEAR','HOME_GMS', 'HOME_AVG', 'HOME_PCT', 'ROAD_GMS',
       'ROAD_AVG', 'ROAD_PCT', 'OVR_GMS', 'OVR_AVG', 'OVR_PCT'
       ]]

attendance_df.head()
```

```python
cities_df.to_csv('sport_city.csv', index=False, header=True)
attendance_df.to_csv('attendance.csv', index=False, header=True)
```
