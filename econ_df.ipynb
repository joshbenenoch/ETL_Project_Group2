```python
# dependencies 
import pandas as pd
```

```python
# Import Demographic data 
econ_df = pd.read_csv('econdataaTest.csv')
econ_df.head(50)
```

```python
# eliminate "None and NaN" values in DataFrame
for strCol in econ_df.columns:
    econ_df[strCol] = econ_df[strCol].str.replace('None', 'NaN')
econ_df.head()
```

```python
# eliminate blanks and replace with 0 
econ_df['2017'].replace({'':'0'},inplace=True)
econ_df
```

```python
# eliminate commas 
econ_df['2017'] = econ_df['2017'].str.replace(',', '')
```

```python
# convert 2017 strings into floats
econ_df['2017'].astype(str).astype(float)
```

```python
# create CSV 
econ_df.to_csv('econ_df.csv')
```
