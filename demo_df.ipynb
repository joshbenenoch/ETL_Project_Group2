```python
# dependencies 
import pandas as pd
```

```python
# Import Demographic data 
demo = pd.read_csv('City_Demographic.csv')
demo.head(50)
```

```python
demo.dtypes
```

```python
demo_df = pd.DataFrame(demo)
```

```python
demo_df['Census 2010 Total Population'] = demo_df['Census 2010 Total Population'].str.replace(',', '')
demo_df['2017 ACS 5-Year Population Estimate'] = demo_df['2017 ACS 5-Year Population Estimate'].str.replace(',', '')
demo_df['Number of Companies'] = demo_df['Number of Companies'].str.replace(',', '')
demo_df['Total housing units'] = demo_df['Total housing units'].str.replace(',', '')
demo_df['Median Household Income'] = demo_df['Median Household Income'].str.replace(',', '')
demo_df['Foreign Born Population'] = demo_df['Foreign Born Population'].str.replace(',', '')
demo_df['Individuals below poverty level'] = demo_df['Individuals below poverty level'].str.replace(',', '', '%')
demo_df['White alone'] = demo_df['White alone'].str.replace(',', '')
demo_df['Black or African American alone'] = demo_df['Black or African American alone'].str.replace(',', '')
demo_df['American Indian and Alaska Native alone'] = demo_df['American Indian and Alaska Native alone'].str.replace(',', '')
demo_df['Asian alone'] = demo_df['Asian alone'].str.replace(',', '')
demo_df['Native Hawaiian and Other Pacific Islander alone'] = demo_df['Native Hawaiian and Other Pacific Islander alone'].str.replace(',', '')
demo_df['Some Other Race alone'] = demo_df['Some Other Race alone'].str.replace(',', '')
demo_df['Two or More Races'] = demo_df['Two or More Races'].str.replace(',', '')
demo_df['Hispanic or Latino (of any race)'] = demo_df['Hispanic or Latino (of any race)'].str.replace(',', '')
demo_df['White alone, Not Hispanic or Latino'] = demo_df['White alone, Not Hispanic or Latino'].str.replace(',', '')
demo_df['Veterans'] = demo_df['Veterans'].str.replace(',', '')

demo_df.head(5)
```

```python
demo_df['Census 2010 Total Population'].astype(str).astype(int)
demo_df['2017 ACS 5-Year Population Estimate'].astype(str).astype(int)
demo_df['Total housing units'].astype(str).astype(int)
demo_df['Median Household Income'].astype(str).astype(int)
demo_df['Foreign Born Population'].astype(str).astype(int)
demo_df['White alone'].astype(str).astype(int)
demo_df['Black or African American alone'].astype(str).astype(int)
demo_df['American Indian and Alaska Native alone'].astype(str).astype(int)
demo_df['Asian alone'].astype(str).astype(int)
demo_df['Native Hawaiian and Other Pacific Islander alone'].astype(str).astype(int)
demo_df['Some Other Race alone'].astype(str).astype(int)
demo_df['Two or More Races'].astype(str).astype(int)
demo_df['Hispanic or Latino (of any race)'].astype(str).astype(int)
demo_df['White alone, Not Hispanic or Latino'].astype(str).astype(int)
demo_df['Veterans'].astype(str).astype(int)
```
